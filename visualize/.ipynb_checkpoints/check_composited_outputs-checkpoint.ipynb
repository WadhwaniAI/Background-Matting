```python
from resize_image import read_image
```

```python
def show_four_images(image1, image2, image3, image4, size=(20, 10)):
    fig, ax = plt.subplots(2, 2, figsize=size)

    ax[0][0].imshow(image1)
    ax[0][0].set_xticks([], [])
    ax[0][0].set_yticks([], [])
    ax[0][0].set_title("Source image")
    
    ax[0][1].imshow(image2)
    ax[0][1].set_xticks([], [])
    ax[0][1].set_yticks([], [])
    ax[0][1].set_title("Source background")

    ax[1][0].imshow(image3)
    ax[1][0].set_xticks([], [])
    ax[1][0].set_yticks([], [])
    ax[1][0].set_title("Target background")
    
    ax[1][1].imshow(image4)
    ax[1][1].set_xticks([], [])
    ax[1][1].set_yticks([], [])
    ax[1][1].set_title("Composited image")

    plt.show()
```

```python
image_original = read_image('../sample_data/input/00026_img.png')
image_back = read_image('../sample_data/input/00026_back.png')
target_back = read_image('../sample_data/background/00026.png')
image_composited = read_image('../sample_data/output/00026_compose.png')
```

```python
show_four_images(image_original, image_back, target_back, image_composited)
```
